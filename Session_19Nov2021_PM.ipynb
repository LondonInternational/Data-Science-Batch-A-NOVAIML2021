```python
# create list of numbers, [7,14,....70] and find out how many are even and how many are odd.

numList=list(range(7,71,7))
print(numList)
cntEven=0
cntOdd=0

for item in numList:
  if  item % 2 == 0:
    cntEven = cntEven + 1
  else:
    cntOdd = cntOdd + 1

print('Even Count is ',cntEven)
print('Odd Count is ',cntOdd)
```

```python
# list of numbers - [1,2,...13], print all numbers except 6 from this list

numList=list(range(1,14)) # step = 1

for item in numList:
  if item != 6:
    print(item,end=',')
```

```python
# 10,2,20,4,30,6,40,...100  
# Built-in Function zip(series1,series2)
# for x,y in zip(range(10,101,10), range(2,21,2)):
#   print(str(x)+','+str(y), end=',')
s1 = 10
s2 = 2
for position in range(1,20):
  if position % 2 == 0:
    print(s2,end=',')
    s2 = s2 + 2
  else:
    print(s1,end=',')
    s1 = s1 + 10
```

```python
num = 1 # start
while(num<20):  # stop
    print(num,end=',')
    num = num + 3 # step
```

```python
# 99,97,95,....>80
num = 99 # start
while(num>80):  # stop
    print(num,end=',')
    num = num - 2 # step
```

```python
# [99,97,95,....]>80  ---> sum --> 
sum = 0
num = 99 
while(num>80):  
    sum = sum + num
    num = num - 2 

print(sum)
```
